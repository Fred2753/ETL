# ETL com Python


## **E**xtract

Extraia a lista de usuário a partir do arquivo clientes.csv.

```python
import pandas as pd
import csv

df = pd.read_csv('clientes.csv', sep=";")
listaClientes = df.values.tolist()
print(listaClientes)
```

## **T**ransform

Verificar o nível do cliente:


>> Iniciante:        oferecer carteira de investimento 100% renda fixa

>> Intermediário:    oferecer investimentos 60% renda fixa e 40% ações

>> Avançado:         oferecer investimentos 70% ações e 30% renda fixa



```python
for i in range(0, len(listaClientes)):
  #Tamanho das listas, lembrando que o indice começa em 0 (zero), a quantidade de itens será sempre o indice + 1.
  tlista = len(listaClientes)

  if listaClientes[i][3] == 'Iniciante':

    mensagem =  "Caro cliente, ofertamos um ótimo investimento de 100% em renda fixa para seu perfil"

  elif listaClientes[i][3] == 'Intermediario':

    mensagem =  "Caro cliente, ofertamos um ótimo investimento de 60% em renda fixa e 40% em ações de fundos imobiliarios para seu perfil"

  elif listaClientes[i][3] == 'Avancado':

    mensagem =  "Caro cliente, ofertamos um ótimo investimento de 70% em renda fixa e 30% em ações de fundos imobiliarios para seu perfil"

  listaClientes[i].insert(tlista, mensagem)
```

```python
header = ['UserId', 'UserNome', 'UserIdade', 'UserNivel', 'UserMsg']
with open('mensagens.csv', 'w', newline= '') as msg:
  writer = csv.writer(msg)
  writer.writerow(header)
  writer.writerows(listaClientes)
  msg.close()
```
